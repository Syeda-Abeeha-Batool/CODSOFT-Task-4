# Task 4 Rock Paper Scissors Game

```python
import random
print("\t\t\t\t\t\tRock-Paper-Scissors Game\n")
print("\n\n")
choice = True
while(choice):
    rounds = int(input("Enter the number of rounds of the game : "))
    score1=0
    score2=0
    for i in range(rounds):
        print("\nROUND ",i+1,"\n")
        print("Player 1\n")
        choice1 = input("Enter your choice : \nPress 'r' for Rock\nPress 'p' for Paper\nPress 's' for Scissors\n")
        n = random.randint(0,3)

        if n==0:
            choice2 = 'r'
        elif n==1:
            choice2 = 'p'
        elif n==2:
            choice2 = 's'

        if choice1 == 'r':
            if choice2 == 'r':
                score1=score1
                score2=score2
                print("Player 1 : rock\nComputer : rock")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 'p':
                score1=score1
                score2=score2+1
                print("Player 1 : rock\nComputer : paper")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 's':
                score1=score1+1
                score2=score2
                print("Player 1 : rock\nComputer : scissors")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
        elif choice1 == 's':
            if choice2 == 'r':
                score1=score1
                score2=score2+1
                print("Player 1 : scissors\nComputer : rock")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 'p':
                score1=score1+1
                score2=score2
                print("Player 1 : scissors\nComputer : paper")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 's':
                score1=score1
                score2=score2
                print("Player 1 : scissors\nComputer : scissors")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
        elif choice1 == 'p':
            if choice2 == 'r':
                score1=score1+1
                score2=score2
                print("Player 1 : paper\nComputer : rock")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 'p':
                score1=score1
                score2=score2
                print("Player 1 : paper\nComputer : paper")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2)
            elif choice2 == 's':
                score1=score1
                score2=score2+1
                print("Player 1 : paper\nComputer : scissors")
                print("\nPlayer 1 Score: ",score1,"\nComputer Score: ",score2) 


    if score1>score2:
        print("\n\nCONGRATULATIONS! You Won....")
    elif score1<score2:
        print("\n\nOops! You Lose....")
    else:
        print("It's a DRAW....")
        
    
    ch = input("Do you want to play another game? (Y/y) or (N/n) ")
    if ch=="Y" or ch=="y":
        choice = True
    else:
        choice = False

print("\nThank you!!") 
```
